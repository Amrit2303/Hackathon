```python
data_train=data_train.drop(['is_timer','campaign_id'],axis=1) # dropping the independent variable from training dataset
data_train.head() 
```

```python
Y = data_train.pop('click_rate') # dropping the target variable
X = data_train 
```

```python
from sklearn.model_selection import train_test_split   # splitting into training and testing dataset
X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size=0.2, random_state=42)
```

```python
from flaml import AutoML
am = AutoML()                # using automl to get the best regression algorithm
```

```python
am.fit(X_train, y_train, task="regression",metric='rmse',time_budget=500)
```

```python
print('Best ML leaner:', am.best_estimator)
```

```python
y_pred = am.predict(data_test)
y_pred[:5]
```

```python
from flaml.ml import sklearn_metric_loss_score  # calculating the error score
print('data_train rmse', '=', sklearn_metric_loss_score('rmse', am.predict(X_train), y_train))
print('data_test rmse', '=', sklearn_metric_loss_score('rmse', am.predict(X_test), y_test))
```
