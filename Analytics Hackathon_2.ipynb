```python
data_train=data_train.drop(['is_timer','campaign_id'],axis=1) # dropping the independent variable from training dataset
data_train.head() 
```

```python
data_test=data_test.drop(['is_timer','campaign_id'],axis=1) # dropping the independent variable from testing dataset
data_test.head() 
```

```python
data_train['times_of_day']=data_train['times_of_day'].replace({'Noon':0,'Morning':1,'Evening':2})
```

```python
data_test['times_of_day']=data_test['times_of_day'].replace({'Noon':1,'Morning':0,'Evening':2})
```

```python
Y = data_train.pop('click_rate') # dropping the target variable
X = data_train 
```

```python
from sklearn.model_selection import train_test_split   # splitting into training and testing dataset
X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python
from flaml import AutoML
sc = AutoML()
```

```python
sc.fit(X_train, y_train, task="regression",metric='rmse',time_budget=400)
```

```python
print('Best ML leaner:', sc.best_estimator)
```

```python
from flaml.ml import sklearn_metric_loss_score
print('train rsme', '=', sklearn_metric_loss_score('rsme', sc.predict(X_train), y_train))
print(' test rsme', '=', sklearn_metric_loss_score('rsme', sc.predict(X_test), y_test))
```

```python
y_pred = sc.predict(data_test)
y_pred[:5]
```

```python
output_dataset=pd.DataFrame(y_pred,columns=['click_rate'])
output_dataset.head()
```
